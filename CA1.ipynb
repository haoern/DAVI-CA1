```python
import os
import pandas as pd
```

```python
# assign path
path, dirs, files = next(os.walk("./Datasets/"))
file_count = len(files)
# create empty list
dataframes_list = []

# append datasets to the list 
for i in range(file_count):
    print(f"./Datasets/{files[i]}")
    temp_df = pd.read_csv(f"./Datasets/{files[i]}")
    dataframes_list.append(temp_df)
    
# display datasets
for dataset in dataframes_list:
    display(dataset)
```

```python
import functools as ft
# property_df = pd.DataFrame()
# property_df.join(dataframes_list[1:4], on='PropertyID')
property_df = ft.reduce(lambda left, right: pd.merge(left, right, on='PropertyID', how='outer'), dataframes_list[1:4])
```

```python
property_df = pd.merge(property_df, dataframes_list[4], left_on='PropertyInSubzone', right_on='SubzoneId', how='left')
```
