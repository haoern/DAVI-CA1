```python
import os
import pandas as pd
```

## “Which residential properties in Singapore provide the best value for money when considering price, accessibility, and nearby amenities?”

```python
# assign path
path, dirs, files = next(os.walk("./Datasets/"))
file_count = len(files)
# create empty list
dataframes_list = []

# append datasets to the list 
for i in range(file_count):
    print(f"./Datasets/{files[i]}")
    temp_df = pd.read_csv(f"./Datasets/{files[i]}")
    dataframes_list.append(temp_df)
    
# display datasets
#for dataset in dataframes_list:
    #display(dataset)
```

```python
dataframes_list[0].dropna(inplace=True)
```

```python
dataframes_list[0]
```

```python
dataframes_list[1].drop_duplicates(inplace=True)
dataframes_list[1] = dataframes_list[1].fillna(0)
dataframes_list[1] = dataframes_list[1].replace("NIL", 0)
dataframes_list[1] = dataframes_list[1].astype(int)
dataframes_list[1]["Ameneties_Count"] = (dataframes_list[1].iloc[:, 1:] != 0).sum(axis=1)
dataframes_list[1]
```

```python
dataframes_list[2]
```

```python
dataframes_list[3].drop_duplicates(inplace=True)
dataframes_list[3].drop(columns=["PropertyNumberOfFloors", "NumberOfBedroomsInTransactedUnit"], inplace=True)
dataframes_list[3] = dataframes_list[3].fillna("Unknown")
```

```python
df = pd.merge(dataframes_list[3], dataframes_list[2], on="PropertyID", how="left")
df.dropna(inplace=True)
df
```

```python
df = pd.merge(df, dataframes_list[1], on="PropertyID", how="left")
df.fillna(0, inplace=True)
```

```python
df['MRT_Score'] = (
    df['NumberOfMRTsWithin500Mts']
    - df['NumberOfMRTsWithin500Mts'].min()
) / (
    df['NumberOfMRTsWithin500Mts'].max()
    - df['NumberOfMRTsWithin500Mts'].min()
)
df['BusStop_Score'] = (
    df['NumberOfBusStopsWithin500Mts']
    - df['NumberOfBusStopsWithin500Mts'].min()
) / (
    df['NumberOfBusStopsWithin500Mts'].max()
    - df['NumberOfBusStopsWithin500Mts'].min()
)
```

```python
df.to_csv("Cleaned Dataset.csv")
```
